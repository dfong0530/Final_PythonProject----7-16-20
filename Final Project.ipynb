```python
import random
import time
import sys

# This is my arcade class. This is the parent class of the games class and restraunt class
class Arcade:
    
    def __init__(self,name,location,cost):
        
        # Variables name --> name of arcade, location of arcade,cost of admittance
        
        self.name = name
        self.location = location
        self.cost = cost
        
    def enter(self):
        
        # This method will be run at the beginning of my games class --> It cost x dollars to enter
        
        print(f"Welcom to {self.name}! The cost of admittance is {self.cost}!")
        
        money = int(input("Enter the amount your going to give to the cashier: "))
        
        while money < self.cost:
            
            print("This is not enough money to enter: ")
            
            print()
            money = int(input("Enter a new amoutnt"))
            
        change = money - self.cost
        
        if change != 0:
            
            print(f"Your change is {change} dollars.")
            
        print("Welcome to the aracade!")
        
     # This method will be run at the end when they are leaving   
    def leave(self):
        
        print("Thank you for comming.")
        print()
        print("We hope to see you soon!")
        
        
        
        
```

```python
class Restraunt(Arcade):
    
    def __init__(self,name,location,cost):
        
        super().__init__(name,location,cost)
        
        print("Hello! Welcome to David's cafe.")
        print("We accepts cash.")
        
        print()
        
        print("We also take credit cards but we don't give them back =)")
        
        self.enter_restraunt()
        
    def enter_restraunt(self):
        
        choice = input("Type 's' to stay at the cafe. Type 'l' to leave the cafe: ")
        
        while choice != 'l':
            
            print()
            # Store menu's in dictionary
            food = {"hot dogs":5,"pizza":4,"chicken wings":8,"fries":3,"salad":9}
            drinks = {"water":1,"fruit punch":3,"coke":3,"vitamin water":4}
            
            decide = input("Type 'f' to see the food menu or type 'd' to see the drinks menu: ")
            
            while decide != 'f' and decide != 'd':
                # Check if they entered somehting invalid
                print("Invalid choice!")
                decide = input("Type f or d: ")
                
            bill = 0
                
            while decide == 'f':
                print()
                print(food)
                
                choose = input("Type in what you would like to order. Make sure to spell correctly: ")
                
                bill += food[choose]
                
                print("This is your bill: " + str(bill))
                
                decide = input("Type f to order more food. Type d to order drinks or type l to leave: ")
                
                
            while decide == 'd':
                print()
                print(drinks)
                
                choose = input("Type in what you would like to order. Make sure you spell correctly: ")
                bill += drinks[choose]
                
                print("This is your bill: " + str(bill))
                
                decide = input("Type d to order more drinks or type l to leave: ")
                
            print()
            
            choice = input("Type 's' to stay or type 'l' to leave: ")
            
        self.leave_restraunt(bill)
        
    def leave_restraunt(self,bill):
        # method --> Call this when they are done with their meal
        print("This is how much you owe: " + str(bill))
        
        decide = input("Type 'cd' to pay with credit card. Type 'c' to pay with cash. Or type 'dd' if you want to dine and dash: ")
        
        while decide != 'cd' and decide != 'c' and decide != 'dd':
            
            print("Invalid choice")
            print("Type cd or c or dd: ")
            
        print()
        
        if decide == 'cd' or decide == 'c':
            
            print("Payment successful.")
            print()
            # method from arcade class
            self.leave()
            
        else:
            # If the customers dine and dash
            print("Just kidding!")
            
            print("The cops are here!")
            
            print()
            
            print("Your going to jail!")
```
